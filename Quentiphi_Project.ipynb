```python
df['Employee_Resignation_Status'] = df['Employee_Resignation_Status'].replace({
    'Yes': 1,
    'No': 0
})
```

```python
features = df[[
    'Performance_Rating',
    'Bonus',
    'Employee_Engagement_Score',
    'Employee_Team_Collaboration_Effectiveness'
]].dropna()

target = df.loc[features.index, 'Employee_Resignation_Status']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
df.loc[features.index, 'Attrition_Risk'] = model.predict(features)
```

```python
df[['Attrition_Risk','Employee_Resignation_Status']].head()
```

```python
from sklearn.metrics import accuracy_score

pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, pred))
```

```python
cols = [
    'Bonus',
    'Annual_Salary_Increase_Percentage',
    'Performance_Bonus_Percentage',
    'Employee_Annual_Salary_Adjustment',
    'Performance_Rating'
]

comp_df = df[cols].copy()
```

```python
for col in cols:
    comp_df[col] = pd.to_numeric(comp_df[col], errors='coerce')

comp_df = comp_df.dropna()
comp_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled = scaler.fit_transform(comp_df)

comp_scaled = pd.DataFrame(scaled, columns=cols, index=comp_df.index)
```

```python
comp_scaled['Fair_Pay_Score'] = (
    0.4 * comp_scaled['Performance_Rating'] +
    0.2 * comp_scaled['Bonus'] +
    0.2 * comp_scaled['Performance_Bonus_Percentage'] +
    0.2 * comp_scaled['Annual_Salary_Increase_Percentage']
)
```

```python
threshold = comp_scaled['Fair_Pay_Score'].quantile(0.25)

comp_scaled['Underpaid'] = comp_scaled['Fair_Pay_Score'] < threshold
```

```python
df.loc[comp_scaled.index, 'Underpaid'] = comp_scaled['Underpaid']
```
